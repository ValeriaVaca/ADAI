### Introducción
La gestión de riesgos y el análisis de eventos adversos constituyen herramientas fundamentales para comprender la dinámica de amenazas que afectan al territorio ecuatoriano. Ecuador, por su ubicación geográfica y características geológicas, climáticas y sociales, está expuesto de manera recurrente a distintos eventos como inundaciones, deslizamientos, incendios forestales, sismos y otros fenómenos que generan impactos significativos sobre la población, infraestructura y actividades económicas. Desde la perspectiva de la Oficina de las Naciones Unidas para la Reducción del Riesgo de Desastres, el riesgo de desastres se entiende como la combinación de la probabilidad de que ocurra un evento adverso y la magnitud de sus consecuencias negativas, en función de la amenaza, la exposición y la vulnerabilidad de un territorio (UNDRR, 2017). Bajo esta lógica, las inundaciones representan una de las amenazas de mayor recurrencia e impacto socioeconómico en el país, particularmente en las provincias de la región Costa.

Entre los fenómenos climáticos que explican la intensificación de los eventos de inundación en Ecuador destaca El Niño-Oscilación del Sur (ENOS), cuando este fenómeno se manifiesta, altera los patrones normales de circulación atmosférica y genera lluvias intensas y prolongadas en la costa ecuatoriana, favoreciendo el desbordamiento de ríos, la saturación de los sistemas de drenaje urbano y, en consecuencia, el incremento de inundaciones y deslizamientos de tierra (Universidad Internacional de Valencia [VIU], 2025).

La relación entre El Niño y las inundaciones en Ecuador está documentada en los episodios más severos de las últimas décadas. El evento de 1982-1983, considerado uno de los más fuertes del siglo XX, ocasionó pérdidas económicas cercanas a los USD 640 millones y una contracción del PIB nacional próxima al 5 %, afectando especialmente a la infraestructura vial, cuya rehabilitación demandó alrededor de USD 210 millones (Teleamazonas, 2026). 

Más recientemente, la temporada lluviosa de 2023, enmarcada en un escenario de El Niño 2023-2024 de magnitud moderada a fuerte, provocó inundaciones de gran magnitud en Chone, en la cuenca media y baja del río Daule y en Esmeraldas. Para ese mismo episodio, el SGR (2023) proyectó la afectación potencial de 488 parroquias en 143 cantones de 17 provincias, identificando 78 cantones con exposición alta y 86 con exposición media al riesgo de inundación, concentrados principalmente en Guayas, Manabí, Los Ríos, El Oro y Esmeraldas.

El acceso a bases de datos históricas de eventos adversos permite estudiar estas inundaciones desde una perspectiva cuantitativa y espacial, facilitando la identificación de patrones de ocurrencia, niveles de afectación y tendencias temporales asociadas a los episodios de El Niño y a la estacionalidad lluviosa de la costa ecuatoriana. Este tipo de análisis es clave para fortalecer la planificación territorial, mejorar la capacidad de respuesta institucional y apoyar la toma de decisiones en gestión del riesgo, especialmente en las zonas históricamente más vulnerables a este tipo de eventos.

En este contexto, el presente trabajo utiliza una base de datos de eventos adversos registrados en Ecuador entre los años 2010 y 2026, con énfasis en los eventos de inundación, con el fin de aplicar técnicas de análisis exploratorio de datos y obtener información relevante sobre el comportamiento de estas inundaciones, su distribución territorial y temporal, y sus impactos asociados sobre la población.

### 3.1. Importación de librerias

### 3.2. Depuración inicial de datos

#### 3.2.1. Corregir tipos de datos

*Al importar los datos originales, se identificaron columnas con tipos de datos mixtos. Para asegurar la correcta carga de todas las variables y prevenir problemas de memoria, se forzó la conversión de todas las columnas a tipo cadena de texto (string).*

```python
#Importar archivo csv a travez de la libreria Pandas

archivo_csv = "Base_Eventos_2010_2026_eventos_catálogo - Base20102025.csv"
df = pd.read_csv(archivo_csv, sep=",",  encoding="utf-8", dtype=str, low_memory=False)  # Lee el archivo CSV asegurando que todas las columnas se carguen como texto (string), utilizando codificación UTF-8 y desactivando la fragmentación de memoria para evitar advertencias de tipos de datos mixtos en columnas grandes.
df.head(10) # Impresión de los 10 primeras filas
```

#### 3.2.2. modificar nombres de columnas

```python
#Filas y columnas

print(f" Filas: {df.shape[0]}")
print(f" Columnas: {df.shape[1]}")
```

```python
# Hacer minúsculas los nombres de las columnas
df.columns = df.columns.str.lower()

# Filtrar columnas de interes en una nueva lista
df_interes = df[["provincia", "canton", "parroquias", "latitud", "longitud", "evento", "causa", "categoria del evento", "fecha del evento", "personas afectadas directamente", "afectadas indirectas", "personas damnificadas"]].copy()

# Función para quitar y remplazar espacios es los titulos
def limpiar_espacios (x):
    x.columns= x.columns.str.replace(" ", "-")
    return x

df_eventos = limpiar_espacios(df_interes)
df_eventos.columns
```

#### 3.2.3. Modificación de tipos de datos | Corregir inconsistencia de datos | Tratar datos nulos

```python
# Debido a la correccion inicial de datos, todos son Dtype .Str 
df_eventos.info()
```

```python
# Conversión de coordenadas a float con funciones
def convertir_float(df, columna):
    df[columna] = df[columna].astype(str).str.replace(",", ".").astype(float)
    return df

df_eventos = convertir_float(df_eventos, "latitud")
df_eventos = convertir_float(df_eventos, "longitud")
df_eventos.head()
```

```python
#Dimension inicial de dataframe
df_eventos.shape
```

```python
#Inconsistencias en los datos numericos
df_eventos["personas-afectadas-directamente"] = df_eventos["personas-afectadas-directamente"].str.replace(".", "")
df_eventos["afectadas-indirectas"] = df_eventos["afectadas-indirectas"].str.replace(".", "")
```

```python
#Función de llenar valores NaN por 0 
def remplazar_nulos(df):
    df = df.fillna(0)
    return df

df_eventos = remplazar_nulos(df_eventos)
df_eventos = df_eventos[df_eventos["provincia"] != 0]
df_eventos.head()
```

```python
#Dimension despues de depuración
df_eventos.shape
```

```python
# Conversión de str a int
def convertir_int(df, columna):
    df[columna] = df[columna].astype(int)
    return df

df_eventos = convertir_int(df_eventos, "personas-afectadas-directamente")
df_eventos = convertir_int(df_eventos, "afectadas-indirectas")
df_eventos = convertir_int(df_eventos, "personas-damnificadas")
df_eventos.head()
```

```python
#Conversión a tipo de dato datetime
df_eventos["fecha-del-evento"] = pd.to_datetime(
    df_eventos["fecha-del-evento"].replace("0", pd.NA),
    errors="coerce",
    dayfirst=True
)

df_eventos.head()
```

```python
df_eventos.head()
```

**Caracterización de los eventos de inundación:**
Tras filtrar exclusivamente los registros clasificados como inundación, se obtuvo un subconjunto compuesto por 10 212 eventos distribuidos en las 24 provincias del Ecuador.
La mayoría de los eventos registrados están asociados a la categoría "Época lluviosa", siendo las lluvias y las lluvias intensas las principales causas reportadas. Esto confirma la estrecha relación entre la ocurrencia de inundaciones y los períodos de mayor precipitación en el país.

**Distribución temporal de las inundaciones:**
Tras ordenar las inundaciones ocurridas se identificó que los años con mayor número de inundaciones es el 2025 (1547 eventos) y 2023 (1325 eventos); los años con menor número de registros fueron 2018 (156 eventos) y 2020 (249 eventos). A su vez, los resultados muestran una tendencia de incremento en el número de inundaciones registradas a partir de 2021, alcanzando los valores máximos durante el período 2023–2025. 

**Distribución espacial de las inundaciones:**
La distribución espacial evidencia una concentración importante de eventos en provincias de la Costa y la región Amazónica. **Impacto de las inundaciones sobre la población:**
Las inundaciones han generado una afectación significativa en la población ecuatoriana. Las provincias con mayor número de personas afectadas directamente fueron Guayas, Manabí, Los Ríos y Esmeraldas, concentrando los mayores impactos a nivel nacional. En cuanto a las personas damnificadas, Manabí registró la cifra más alta, seguida de Los Ríos, Guayas y Esmeraldas. Por otro lado, los mayores valores de afectados indirectos se presentaron en Pichincha, Azuay y Pastaza, evidenciando que los efectos de las inundaciones trascienden las zonas directamente impactadas y afectan a un número considerable de habitantes mediante interrupciones de servicios, movilidad e infraestructura.

**Eventos extremos registrados:** 
El evento de inundación con mayor número de personas afectadas directamente ocurrió en el cantón Chone, provincia de Manabí, el 21 de febrero de 2024, asociado a lluvias intensas, registrando 47800 personas afectadas. Respecto a los afectados indirectos, el valor máximo se registró en Sangolquí, provincia de Pichincha, el 11 de febrero de 2022, con 73000 personas afectadas indirectamente. Se evidencia la presencia de valores atípicos en las tres variables analizadas: personas afectadas directamente, personas afectadas indirectamente y personas damnificadas. En comparación, los eventos clasificados como Naturales presentan una frecuencia significativamente menor, mientras que los eventos de origen Antrópico constituyen una proporción reducida del total.
Estos resultados evidencian que la ocurrencia de inundaciones en Ecuador está estrechamente relacionada con los periodos de mayor precipitación, lo que confirma el carácter estacional de este tipo de amenazas.

**Evolución temporal de los eventos de inundación:**
El análisis temporal muestra una variación considerable en el número anual de eventos registrados entre 2010 y 2026. Durante el período comprendido entre 2010 y 2022 se observan fluctuaciones importantes, con valores que oscilan aproximadamente entre 150 y 620 eventos por año.

**Correlación entre afectados directos, afectados indirectos y damnificados:**
La correlación entre afectados directos y afectados indirectos es prácticamente nula, no hay ninguna relación lineal entre ambas.

La correlación entre afectados indirectos y damnificados es prácticamente nula, tampoco hay relación lineal.

Además, se comprobó que su ocurrencia está estrechamente asociada a la época lluviosa y a la presencia de lluvias intensas, lo que evidencia el carácter predominantemente hidrometeorológico de este fenómeno.

La distribución espacio-temporal de las inundaciones reveló que su incidencia no es homogénea en el territorio nacional. A partir de 2023 se observó un incremento considerable en el número de eventos registrados, mientras que las provincias de Guayas, Manabí, Los Ríos, Esmeraldas y El Oro concentraron cerca del 59 % de las inundaciones del país. Este comportamiento evidencia que las zonas costeras constituyen las áreas con mayor recurrencia e impacto de este tipo de amenaza.

El análisis de la afectación poblacional mostró que las provincias con mayor frecuencia de inundaciones también concentran el mayor número de personas afectadas directamente, destacándose Guayas, Manabí, Los Ríos y Esmeraldas. Sin embargo, la distribución de los afectados indirectos y de los damnificados presenta diferencias entre provincias, lo que demuestra que el impacto de una inundación depende no solo de la cantidad de eventos registrados, sino también de las características sociales, territoriales y de infraestructura de cada localidad.

La identificación de valores atípicos en las variables de afectación confirma la existencia de eventos extremos capaces de generar consecuencias significativamente superiores al comportamiento habitual de la serie. Casos como las inundaciones ocurridas en Chone (2024), Sangolquí (2022) y Esmeraldas (2023) evidencian que, aunque la mayoría de las inundaciones producen afectaciones moderadas, existen eventos excepcionales con un elevado impacto sobre la población.

En conjunto, el análisis exploratorio permitió caracterizar el comportamiento espacial, temporal y poblacional de las inundaciones en Ecuador, identificando patrones y tendencias que constituyen una base confiable para el desarrollo de análisis geoespaciales y modelos de evaluación de susceptibilidad y riesgo, facilitando la toma de decisiones orientadas a la gestión integral del riesgo de desastres.


## <font color="purple">8. Recomendaciones</font>

Se recomienda realizar un proceso de estandarización y limpieza de los datos antes de iniciar cualquier análisis exploratorio o estadístico. En particular, es conveniente convertir inicialmente las variables al tipo de dato string (str) para facilitar la detección y corrección de inconsistencias, valores mixtos y errores de formato, garantizando posteriormente la asignación del tipo de dato adecuado para cada variable.

Es recomendable implementar un procedimiento sistemático de control de calidad de los datos, que incluya la identificación y eliminación de registros duplicados, el tratamiento de valores nulos, la corrección de tipos de datos y la estandarización de los formatos de las variables. Estas acciones mejoran la confiabilidad de la información y permiten obtener resultados más precisos en el análisis de tendencias y patrones de los eventos adversos.

La metodología de depuración y preprocesamiento aplicada en este estudio puede emplearse en otras bases de datos históricas relacionadas con la gestión del riesgo de desastres en el Ecuador, ya que este tipo de información suele presentar inconsistencias derivadas de diferentes procesos de registro y captura de datos. Su aplicación favorece la generación de bases de datos más consistentes y aptas para análisis estadísticos, geoespaciales y de modelamiento.
